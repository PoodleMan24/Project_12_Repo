2. Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the `BusinessTypeID` and `BusinessType` fields.

```python
# Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the BusinessTypeID and BusinessType fields
projection = {"BusinessTypeID": 1, "BusinessType": 1, "_id": 0}
query = {"BusinessType": "Restaurant/Cafe/Canteen"}

results = establishments.find(query, projection).limit(5)
for result in results:
    print(result)
```

5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude
for document in establishments.find({}):
    # Convert latitude and longitude strings to Decimal numbers
    latitude_decimal = float(document['geocode']['latitude'])
    longitude_decimal = float(document['geocode']['longitude'])
    
    # Update the document with the new latitude and longitude values
    establishments.update_one(
        {'_id': document['_id']},
        {'$set': {'geocode.latitude': latitude_decimal, 'geocode.longitude': longitude_decimal}}
    )
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Change the data type from String to Integer for RatingValue
for document in establishments.find({}):
    if 'RatingValue' in document and document['RatingValue'] is not None:
        try:
            rating_value_int = int(document['RatingValue'])
        except ValueError:
            # Skiping documents where RatingValue is not a valid integer
            print(f"Skipping document with ID {document['_id']}: RatingValue is not a valid integer.")
            continue

        establishments.update_one(
            {'_id': document['_id']},
            {'$set': {'RatingValue': rating_value_int}}
        )
    else:
        # Skiping documents where RatingValue is None
        print(f"Skipping document with ID {document['_id']}: RatingValue is None.")
```

```python
# Check that the coordinates and rating value are now numbers
establishments.find_one({})
```
